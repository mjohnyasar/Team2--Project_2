## 2. Most common 15 Name or Surname of killed people 

```python
kill.head()
```

```python
kill.name.value_counts()
```

```python
# Problem: TK TK is most probably not a name or surname. We will clean them.
pairs = kill.name[(kill.name != 'TK TK') & (kill.name != "TK Tk")].str.split()
pairs.head()
```

**INSTRUCTION : Plot Most common 15 Name or Surname of killed people** 

```python
pairs
```

```python
len(pairs)
```

```python
pairs.isnull().sum()
```

```python
#index i yeniden sıralamak istediğimizde reset_index ie yeni index tanımlayıp 
# öncekini silmeli(drop)  ve mevcut dosya üzerine yazmalıyız (inplace=True)

pairs.reset_index(inplace=True, drop=True)
pairs
```

```python
L=[]
for i in range(len(pairs)):
    L.append(len(pairs[i]))
pd.DataFrame(L).value_counts()
```

```python
#  en fazla 4 lü isim var
```

```python
N2,N3,N4=[],[],[]

for i in range(len(pairs)):
    if    len(pairs[i])==2:   N2.append(pairs[i])         
    elif  len(pairs[i])==3:   N3.append(pairs[i])  
    else :  N4.append(pairs[i])
```

```python
a,b=zip(*N2);  c,d,e=zip(*N3);  f,g,h,i=zip(*N4)
```

```python
names=a+c+d+f+g+h
surnames=b+e+i
allnames=names+surnames
```

```python
mc_n = Counter(names).most_common(15)
mc_s = Counter(surnames).most_common(15)
mc_a = Counter(allnames).most_common(15)
```

```python
df_n=pd.DataFrame(mc_n,columns = ["names", "commons"])
df_s=pd.DataFrame(mc_s,columns = ["names", "commons"])
df_a=pd.DataFrame(mc_a,columns = ["names", "commons"])
k=[df_n,df_s,df_a]
L=("Most common 15 NAMES of killed people", 
   "Most common 15 SURNAMES of killed people", 
   "Most common 15 Names and Surnames of killed people")
```

```python

for i in range(len(k)):
    fig = plt.figure(figsize = (12,5))

    sb.set_context( rc ={'xtick.labelsize': 8.8})
    sb.set(style = "white", palette = "muted")
    sb.barplot(x = "names", y ="commons",data= k[i])
    plt.title(L[i], fontsize=16)
    plt.xlabel("Name or surname of killed people ", fontsize=14 )
    plt.ylabel("Frequency", fontsize=14)
    plt.xticks(rotation = 45, fontsize = 14)
    
    for index, row in k[i].iterrows():
        plt.text(row.name,row.commons, round(row.commons,1), color='black', ha="center", fontsize=16)
  
    plt.tight_layout()
plt.show()
```

```python

#istenen grafik örneği
```

Problem: missing value on race column.

**INSTRUCTION-1: Drop the all raws that contain missing value (dropna)**

```python
kill.isnull().sum()
```

```python
kill.dropna(subset = ["race"], inplace = True)
kill.shape
```

**INSTRUCTION-2: Demonstrate Race Ratio of Killed People by a Pie Chart**

```python
krvc=kill.race.value_counts(normalize=True)
krvc
```

```python
krvd=prc(krvc.to_frame())
krvd
```

```python
krvc.index
```

```python
fig = plt.figure(figsize = (20, 10))

colors = ["grey",'blue','red','yellow','green',"maroon"]
plt.title("Killed People Acording To Races", fontsize=14, color="b")
plt.pie(krvd.race, labels =krvd.index,autopct='%1.1f%%', colors = colors)
plt.legend(loc="upper right")
plt.show()
```

```python
fig = plt.figure(figsize = (20, 10))
colors  = ("grey", "blue", "red", "yellow", "green", "brown")
krvd.race.plot.pie( autopct='%1.1f',  colors = colors,) #subplots=True, 
plt.legend(loc="upper right")
plt.title("Killed People According to Races",fontsize=16, color="blue")
plt.show()
```
